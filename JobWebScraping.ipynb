```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

Start by performing a GET request on the url above and convert the response into a BeautifulSoup object.

```python
URL = 'https://realpython.github.io/fake-jobs/'
response = requests.get(URL)
if response.status_code == requests.codes.ok:
    soup = BeautifulSoup(response.text, features="html.parser")
    print(soup.prettify())
else:
    response.raise_for_status()
```

Use the .find method to find the tag containing the first job title ("Senior Python Developer"). Hint: can you find a tag type and/or a class that could be helpful for extracting this information? Extract the text from this title.

```python
soup.find('h2').text
```

Now, use what you did for the first title, but extract the job title for all jobs on this page. Store the results in a list.

```python
jobs = soup.findAll('h2')
jobs_list = []
for job in jobs:
    jobs_list.append(job.text.strip())
jobs_list
```

```python
companies = soup.findAll('h3')
companies_list = []
for com in companies:
    companies_list.append(com.text.strip())
companies_list
```

```python
location = soup.findAll('p', class_="location")
location_list = []
for loc in location:
    location_list.append(loc.text.strip())
location_list
```

```python
date_posted = soup.findAll('time')
date_posted_list = []
for date in date_posted:
    date_posted_list.append(date.text.strip())
date_posted_list
```

Take the lists that you have created and combine them into a pandas DataFrame.

```python
jobs_df = (
    pd.DataFrame([jobs_list, companies_list, location_list, date_posted_list])
    .transpose()
    .rename(columns={0:'Job',1:'Company',2:'Location',3:'Date_Posted'})
)
jobs_df
```

```python
jobs = jobs_df['Job'].to_list()
job_links = []
for i in range(len(jobs)):
    job_links.append('https://realpython.github.io/fake-jobs/jobs/' + '-'.join(jobs[i].lower().replace('/',' ').replace(',', ' ').replace('(','').replace(')','').split()) + '-' + str(i) + '.html')
jobs_df['Job_Links'] = job_links
jobs_df
```

3. a. Start by looking at the page for the first job, https://realpython.github.io/fake-jobs/jobs/senior-python-developer-0.html. Using BeautifulSoup, extract the job description paragraph.

```python
url = jobs_df['Job_Links'].iloc[0]
response = requests.get(url)
if response.status_code == requests.codes.ok:
    soup = BeautifulSoup(response.text, features="html.parser")
else:
    response.raise_for_status()

paragraphs = soup.findAll('p')
longest_paragraph = max(paragraphs, key=lambda x: len(x.text))
longest_paragraph.text.strip()
```

b. We want to be able to do this for all pages. Write a function which takes as input a url and returns the description text on that page. For example, if you input "https://realpython.github.io/fake-jobs/jobs/television-floor-manager-8.html" into your function, it should return the string "At be than always different American address. Former claim chance prevent why measure too. Almost before some military outside baby interview. Face top individual win suddenly. Parent do ten after those scientist. Medical effort assume teacher wall. Significant his himself clearly very. Expert stop area along individual. Three own bank recognize special good along.".

```python
def description_of_job_url(url):
    response = requests.get(url)
    if response.status_code == requests.codes.ok:
        soup = BeautifulSoup(response.text, features="html.parser")
    else:
        response.raise_for_status()
    return max(soup.findAll('p'), key=lambda x: len(x.text)).text.strip()
    #Finds the longest paragraph in the URL and returns it as the description of the job

description_of_job_url(jobs_df['Job_Links'].iloc[8])
```

c. Use the .apply method on the url column you created above to retrieve the description text for all of the jobs.

```python
jobs_df['Job_Description'] = jobs_df['Job_Links'].apply(description_of_job_url)
jobs_df
```
